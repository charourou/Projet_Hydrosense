### Recherche par département

Comment chercher les piezos qui ne sont plus d'actualités ?N

```python
print("\n--- Recherche par département 79 ---")
s79 = Entite().rechercher_stations(code_dep="79")

collecteur = GestionnairePiezometrie(dossier_sortie= './raw_data')
```

```python
derniere_collecte = []

for nom_piezo in s79[:]:
    df = collecteur.telecharger_forage(nom_piezo)
    if not type(df) == bool:
        derniere_collecte.append(df['date_mesure'].iloc[-1])
    else:
        s79.remove(nom_piezo)
```
